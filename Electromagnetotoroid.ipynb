```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters for our torus
a = 1  # Radius of the tube
c = 2     # Radius from the center to the center of the tube (m)

# Generate the toroidal coordinates
phi = np.linspace(0, 2 * np.pi, 100)
theta = np.linspace(0, 2 * np.pi, 100)
phi, theta = np.meshgrid(phi, theta)

# Convert toroidal coordinates to Cartesian coordinates
x = (c + a*np.cos(phi)) * np.cos(theta)
y = (c + a*np.cos(phi)) * np.sin(theta)
z = a * np.sin(phi)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, color='c', edgecolor='k', linewidth=0.5, alpha=0.6)

# Adding axis labels with adjusted positioning
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z', labelpad=2)  # Adjusting labelpad to move Z label outwards

ax.set_title("Torus in Toroidal Coordinates")
plt.show()

# Gravitational mass equation
a = 1.0e-1
c=  5.0
mu = 4.e-7*np.pi    # vacuum permeability
S = 1
I_prime = 1.0e3         # A
Ib = 1.0e3
rho = 1.602e-2          # Plasma density in the current inside toroid
mi = 1.0e3              # initial mass (kg)
N = 6                   # Number of units working in the machine
                        # mg transformed mass
mg = mi - N*(mu * 2* np.pi * I_prime *c) / (a**2 * rho)  # multiply by N units

print("Gravitational mass (mg): {:.2f} kg".format(mg))
delta_m=mg-mi
print("mass variation (kg): {:.2f} kg".format(delta_m))

#################################################################

import numpy as np
import matplotlib.pyplot as plt

# Constants
c = 5.0
mu = 4.e-7 * np.pi    # vacuum permeability
S = 1
I_prime = 1.0e3       # A
Ib = 1.0e3
rho = 1.602e-2        # Plasma density in the current inside toroid
mi = 1.0e3            # initial mass (kg)
N = 6                 # Number of units working in the machine

# Range of 'a' and 'b' values
a = np.linspace(0.05, 0.15, 100)
b = np.linspace(1.0, 10.0, 100)

A, B = np.meshgrid(a, b)

# Gravitational mass
mg = mi - N * (mu * 2 * np.pi * I_prime * c) / (A**2 * rho)

# Check the range of mg values
print(f"Max of mg: {np.max(mg)}")
print(f"Min of mg: {np.min(mg)}")


# Plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(A, B, mg, cmap='viridis')

ax.set_xlabel('a')
ax.set_ylabel('b')
ax.set_zlabel('mg', labelpad=8)  # Increase the padding for the z label
ax.set_title('Gravitational mass (mg) as a function of a and b')

# Set z-axis limits based on mg values
ax.set_zlim(np.min(mg), np.max(mg))

fig.colorbar(surf, ax=ax, pad=0.12)

plt.show()

#################################################

import numpy as np
import matplotlib.pyplot as plt

# Toroidal coordinates
theta = np.linspace(0, 2.0*np.pi, 100)  # Poloidal angle
phi = np.linspace(0, 2.0*np.pi, 100)    # Toroidal angle
theta, phi = np.meshgrid(theta, phi)
a, c = 0.2, 1  # Small and large radii
x = (c + a*np.cos(theta)) * np.cos(phi)
y = (c + a*np.cos(theta)) * np.sin(phi)
z = a * np.sin(theta)

# Create the torus using surface plot
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, color='c', edgecolor='k', linewidth=0.5, alpha=0.6)

# Compute the tangential vectors for current density K along the torus surface
# Here, we'll consider the current flowing in the phi direction (toroidal direction)
u = -np.sin(phi)  # -sin(phi) is the x-component of the tangential vector
v = np.cos(phi)   # cos(phi) is the y-component of the tangential vector
w = np.zeros_like(phi)  # z-component is 0 as the current is in the xy-plane

# Draw the quiver (arrows) to represent current density K on the torus
skip = 10  # Skip some points to make the plot less cluttered
ax.quiver(x[::skip, ::skip], y[::skip, ::skip], z[::skip, ::skip],
          u[::skip, ::skip], v[::skip, ::skip], w[::skip, ::skip],
          length=0.1, normalize=True, color='r')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z', labelpad=1)
ax.zaxis.set_rotate_label(False)  # Disable automatic rotation
ax.set_zlabel('Z', rotation=0)
ax.set_title("Torus with Current Density K")

# Adjust the view angle for better visualization
ax.view_init(30, 30)

plt.tight_layout()
plt.show()

###################################################

import numpy as np
import matplotlib.pyplot as plt

# Toroidal coordinates
theta = np.linspace(0, 2.0*np.pi, 100)  # Poloidal angle
phi = np.linspace(0, 2.0*np.pi, 100)    # Toroidal angle
theta, phi = np.meshgrid(theta, phi)
a, c = 0.2, 1  # Small and large radii
x = (c + a*np.cos(theta)) * np.cos(phi)
y = (c + a*np.cos(theta)) * np.sin(phi)
z = a * np.sin(theta)

# Create the torus using surface plot
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, color='c', edgecolor='k', linewidth=0.5, alpha=0.6)

# Compute the tangential vectors for current density K along the torus surface
# Considering the current flowing in the theta direction (poloidal direction)
u = -np.sin(theta) * np.cos(phi)
v = -np.sin(theta) * np.sin(phi)
w = np.cos(theta)

# Draw the quiver (arrows) to represent current density K on the torus
skip = 10  # Skip some points to make the plot less cluttered
ax.quiver(x[::skip, ::skip], y[::skip, ::skip], z[::skip, ::skip],
          u[::skip, ::skip], v[::skip, ::skip], w[::skip, ::skip],
          length=0.1, normalize=True, color='r')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z', labelpad=1)
ax.zaxis.set_rotate_label(False)  # Disable automatic rotation
ax.set_zlabel('Z', rotation=0)
ax.set_title("Torus with Current Density K in Poloidal Direction")

# Adjust the view angle for better visualization
ax.view_init(30, 30)

plt.tight_layout()
plt.show()

```

Note about rho: In a discharge plasma, the charge density, , can vary significantly based on the type of discharge, pressure, and other conditions. However, a typical value for the electron density ne in many common plasmas is on the order of  1.e12 to 1.0e16 electrons per cubic centimeter.

To convert this electron density to charge density , you would use the formula:
=ne e where e is the elementary charge (1.602e-19 Coulombs).

As an example, if \( n_e = 10^{14} \) cm\(^{-3}\), then:
\[
\rho = 10^{14} \times 1.602 \times 10^{-19} = 1.602 \times 10^{-5}
\]
Coulombs per cubic centimeter or \( 1.602 \times 10^{-2} \) Coulombs per cubic meter.
