```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Define the hybrid model
def hybrid_model(y, t):
    # Unpack state variables
    y1 = y[0]
    y2 = y[1]

    # Calculate derivatives
    dy1dt = y2
    dy2dt = -2*y2 - 2*y1

    # Pack derivatives
    dydt = [dy1dt, dy2dt]
    return dydt

# Simulation parameters
y1_0 = 1
y2_0 = 0
y0 = [y1_0, y2_0]
t = np.linspace(0, 10, 1000)

# Solve the differential equation using the hybrid model
sol = odeint(hybrid_model, y0, t)
y1 = sol[:, 0]
y2 = sol[:, 1]

# Plot the results
plt.plot(t, y1, label='y')
plt.plot(t, y2, label='dy/dt')
plt.xlabel('Time (s)')
plt.ylabel('Value')
plt.legend()
plt.show()
```
