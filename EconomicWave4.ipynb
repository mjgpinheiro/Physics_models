```python
import numpy as np
import matplotlib.pyplot as plt
# Number of producers
num_producers = 5

# Time steps
time_steps = 100

# Initialize arrays to store innovation, demand, productivity, and income
innovation = np.zeros((num_producers, time_steps))
demand = np.zeros((num_producers, time_steps))
productivity = np.zeros((num_producers, time_steps))
income = np.zeros((num_producers, time_steps))

# Set initial values for innovation, demand, productivity, and income
innovation[:, 0] = np.random.rand(num_producers)
demand[:, 0] = np.random.rand(num_producers)
productivity[:, 0] = np.random.rand(num_producers)
income[:, 0] = np.random.rand(num_producers)

# Set parameters for the model
innovation_rate = 0.05 #0.05
demand_rate = 0.120 #0.1
productivity_rate = 0.091 #0.1
income_rate = 0.10 #0.1

for t in range(1, time_steps):
    for p in range(num_producers):
        # Update innovation
        innovation[p, t] = innovation[p, t-1] + innovation_rate * np.random.rand()
        
        # Update demand
        demand[p, t] = demand[p, t-1] + demand_rate * (innovation[p, t] - demand[p, t-1])
        
        # Update productivity
        productivity[p, t] = productivity[p, t-1] + productivity_rate * (demand[p, t] - productivity[p, t-1])
        
        # Update income
        income[p, t] = income[p, t-1] + income_rate * (productivity[p, t] - income[p, t-1])
plt.figure(figsize=(12, 6))
plt.plot(np.sum(income, axis=0), label='Total Income')
plt.xlabel('Time Steps')
plt.ylabel('Income')
plt.title('Economic Wave')
plt.legend()
plt.show()
```
