```python
import numpy as np
import matplotlib.pyplot as plt

# Define fundamental constants
h = 1.0546e-34  # Planck's constant (J s)
m = 6.6465e-27  # Mass of helium-4 atom (kg)
kappa = h / m   # Fundamental circulation quantum (m^2/s)

# Define a function to generate a vortex with quantized circulation
def vortex(x, y, x0, y0, n):
    r = np.sqrt((x - x0)**2 + (y - y0)**2)
    theta = np.arctan2(y - y0, x - x0)
    phi = n * theta
    return kappa * np.array([-np.sin(phi), np.cos(phi)])

# Generate a grid of points
N = 301
x = np.linspace(-1, 1, N)
y = np.linspace(-1, 1, N)
X, Y = np.meshgrid(x, y)

# Generate a vortex with quantized circulation
n = 200
vx, vy = vortex(X, Y, 0, 0, n)

# Plot the velocity field and circulation
fig, ax = plt.subplots(figsize=(8, 6))
ax.quiver(X, Y, vx, vy, scale=10)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title(f'Vortex with n={n} (circulation = {n*kappa:.2e} m^2/s)')
plt.show()
```
