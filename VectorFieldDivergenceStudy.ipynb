```python
import numpy as np
import matplotlib.pyplot as plt

# Definir o grid espacial
x, y = np.meshgrid(np.linspace(-5, 5, 10), np.linspace(-5, 5, 10))

# Definir o campo elétrico E = (E_x, E_y)
E_x = x/np.sqrt(x**2 + y**2)
E_y = y/np.sqrt(x**2 + y**2)

# Calculando as derivadas parciais usando a função np.gradient
dE_x_dx, dE_x_dy = np.gradient(E_x, axis=(1, 0))
dE_y_dx, dE_y_dy = np.gradient(E_y, axis=(1, 0))

# Calculando a divergência
div_E = dE_x_dx + dE_y_dy

# Fazendo um plot do campo vetorial
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.quiver(x, y, E_x, E_y, scale=20)
plt.title('Campo Elétrico')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.xlim([-5,5])
plt.ylim([-5,5])

# Fazendo um plot da divergência
plt.subplot(1, 2, 2)
plt.imshow(div_E, extent=(-5, 5, -5, 5), origin='lower', cmap='RdBu_r')
plt.title('Divergência do Campo Elétrico')
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar(label='Divergência')
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

plt.tight_layout()
plt.show()
```
