```python
import numpy as np
import matplotlib.pyplot as plt

def kitaev_chain_hamiltonian(N=10, t=1, Delta=1, mu=0):
    """Construct the Kitaev chain Hamiltonian."""
    # Kinetic term + potential term
    H_0 = -mu * np.eye(N) - t * np.diag(np.ones(N-1), 1) - t * np.diag(np.ones(N-1), -1)
    # Superconducting term
    Delta_matrix = Delta * np.diag(np.ones(N-1), 1) - Delta * np.diag(np.ones(N-1), -1)

    H = np.block([[H_0, Delta_matrix], [Delta_matrix.T.conj(), -H_0]])
    return H

def plot_energy_spectrum(H):
    """Plot the energy spectrum of the Hamiltonian."""
    energies = np.linalg.eigvalsh(H)
    plt.figure(figsize=(8, 6))
    plt.plot(energies, 'o')
    plt.title('Energy Spectrum of the Kitaev Chain')
    plt.xlabel('State Index')
    plt.ylabel('Energy')
    plt.grid(True)
    plt.show()

N = 50  # Number of sites
H = kitaev_chain_hamiltonian(N=N)
plot_energy_spectrum(H)
```
