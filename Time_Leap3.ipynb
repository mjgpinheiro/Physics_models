```python
# Importar as bibliotecas necessrias
import numpy as np
import matplotlib.pyplot as plt

# Parmetros
gamma_r = 0.1  # Parte real de gamma
gamma_i = 0.5  # Parte imaginria de gamma
gamma = gamma_r + 1j * gamma_i  # Gamma complexo

# Intervalo de tempo
t_eval = np.linspace(0, 100, 1000)

# Informao exponencial com gamma complexo
I_exp = np.exp(gamma * t_eval)

# Tempo percebido (logaritmo da informao)
t_percebido_real = np.log(np.abs(I_exp)) / gamma_r
t_percebido_imag = np.angle(I_exp) / gamma_i

# Plotar os resultados
plt.figure(figsize=(12, 6))

# Plotar a informao exponencial (mdulo)
plt.plot(t_eval, np.abs(I_exp), label="Informao Exponencial |I(t)|", color='b')
plt.xlabel("Tempo Real $t$")
plt.ylabel("Informao $|I(t)|$")
plt.yscale('log')  # Usar escala logartmica para a informao
plt.title("Crescimento Exponencial vs Saturao Logartmica com Gamma Complexo")
plt.grid(True)
plt.legend()

# Plotar o tempo percebido real
plt.plot(t_eval, t_percebido_real, label="Tempo Percebido Real $t'_r(t) = \ln|I|/\gamma_r$", color='r')

# Plotar o tempo percebido imaginrio
plt.plot(t_eval, t_percebido_imag, label="Tempo Percebido Imaginrio $t'_i(t) = \mathrm{angle}(I)/\gamma_i$", color='g')

plt.xlabel("Tempo Real $t$")
plt.ylabel("Tempo Percebido $t'(t)$")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
