```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
beta = 1  # Fixed value for beta

# Range of entropy values
S = np.linspace(0.1, 10, 400)

# Different values of gamma
gamma_values = [0.5, 1, 2, 5]

# Plotting
plt.figure(figsize=(10, 6))

for gamma in gamma_values:
    t_prime = (1 / (2 * gamma)) * np.log(S) - (1 / (2 * gamma)) * np.log(beta)
    plt.plot(S, t_prime, label=f' = {gamma}')

plt.xlabel('Entropy (S)')
plt.ylabel('Perceived Time (t\')')
plt.title('Perceived Time t\' as a function of Entropy S for different  values')
plt.legend()
plt.grid(True)
plt.ylim(-3, 3)  # Adjust y-axis limits for better visualization
plt.show()
```
