```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros do sistema
r_toro = 1.0  # Raio maior do toro
r_tubo = 0.3  # Raio menor do tubo do toro

# Definindo o campo toroidal (anapolo)
theta = np.linspace(0, 2*np.pi, 100)
phi = np.linspace(0, 2*np.pi, 100)
theta, phi = np.meshgrid(theta, phi)

# Coordenadas toroidais
x = (r_toro + r_tubo*np.cos(theta)) * np.cos(phi)
y = (r_toro + r_tubo*np.cos(theta)) * np.sin(phi)
z = r_tubo * np.sin(theta)

# Campo toroidal (simples representao, no um clculo real de campo)
Bx = -np.sin(phi)
By = np.cos(phi)
Bz = np.zeros_like(theta)

# Plotando a configurao do campo
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.quiver(x, y, z, Bx, By, Bz, length=0.1, normalize=True)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Simulao de Campo Anapolo em uma Configurao Toroidal')
plt.show()
```
