```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt

# Define constants
m_h = 125  # Higgs boson mass in GeV/c^2
v = 246  # Vacuum expectation value in GeV/c^2
g = 1.2  # Coupling constant

# Define the scalar field potential
def potential(phi):
    return (m_h**2/2)*phi**2 + (g**2/4)*phi**4

# Define the field values for the potential
phi = np.linspace(-v, v, 1000)
V = potential(phi)

# Plot the scalar field potential
fig, ax = plt.subplots()
ax.plot(phi, V)
ax.set_xlabel(r'$\phi$ (GeV/c$^2$)')
ax.set_ylabel(r'$V(\phi)$ (GeV$^4$/c$^8$)')
ax.set_title('Scalar Field Potential')

# Simulate symmetry breaking by finding the minimum of the potential
min_phi = np.argmin(V)
min_V = np.min(V)
ax.scatter(phi[min_phi], min_V, color='r')

# Calculate the Higgs boson mass
m_h_calculated = np.sqrt(2)*g*v
print("Calculated Higgs boson mass: {:.2f} GeV/c^2".format(m_h_calculated))

# Plot the Higgs boson potential energy
higgs_V = potential(m_h_calculated/np.sqrt(2))
ax.axhline(y=higgs_V, color='g', linestyle='--')
ax.text(phi[0], higgs_V+0.1*higgs_V, 'Higgs boson mass', color='g')

# Calculate the critical temperature for symmetry restoration
T_c = m_h_calculated/(np.sqrt(2)*g)
print("Critical temperature for symmetry restoration: {:.2f} GeV".format(T_c))

# Plot the critical temperature
ax.axvline(x=T_c, color='b', linestyle='--')
ax.text(T_c+0.1*T_c, V[0], 'Critical temperature', color='b')

plt.show()
```
