```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
R_s = 1.0  # Schwarzschild radius
Lambda = 0.05  # Planck-scale cutoff
r = np.linspace(R_s + 1e-3, 2*R_s, 500)

# Entropy functions
S_classical = np.log((r - R_s)/(r + R_s))
S_reg = np.log((r - R_s + Lambda*np.exp(-(r-R_s)/Lambda))/(r + R_s))

# Plotting
plt.figure(figsize=(8,6))
plt.plot(r/R_s, S_classical, 'r--', label='Classical $S(r)$')
plt.plot(r/R_s, S_reg, 'b-', label='Regularized $S_{reg}(r)$')
plt.axvline(1, color='k', linestyle=':')
plt.xlabel('$r/R_s$', fontsize=14)
plt.ylabel('Entropy $S/R$', fontsize=14)
plt.title('Entropy Regularization Near $R_s$', fontsize=16)
plt.legend(fontsize=12)
plt.grid(True)

# Inset for Planck region
ax_inset = plt.axes([0.5, 0.3, 0.35, 0.35])
ax_inset.plot(r/R_s, S_reg, 'b-')
ax_inset.set_xlim(1, 1 + 5*Lambda/R_s)
ax_inset.set_title('Planck-scale Transition')
ax_inset.set_xticks([1, 1 + 2.5*Lambda/R_s, 1 + 5*Lambda/R_s])
ax_inset.grid(True)

plt.savefig('entropy_profile.pdf', bbox_inches='tight')
```
