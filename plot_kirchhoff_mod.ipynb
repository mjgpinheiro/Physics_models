```python
!pip install matplotlib
from IPython import get_ipython
from IPython.display import display
# %%
!pip install matplotlib
import matplotlib.pyplot as plt

# Frequency response
f_classical = 5.03  # MHz
f_conformal = 4.82
f_experiment = 4.85

# Q-factor
q_classical = 15.8
q_conformal = 16.6
q_experiment = 16.4
q_error = 0.3

# Create subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))

# Subplot 1: Frequency shift
ax1.axvline(f_classical, color='k', linestyle=':', label='Classical $f_0$')
ax1.axvline(f_conformal, color='r', linestyle='--', label='Theoretical (This work)')
ax1.axvline(f_experiment, color='b', marker='o', label='Experimental (Paez et al., 2014)')
ax1.set_xlabel('Frequency (MHz)')
ax1.set_title('Resonant Frequency Shift ($\kappa=0.05$)')
ax1.legend()

# Subplot 2: Q-factor
ax2.bar(['Classical', 'This work', 'Experiment'],
        [q_classical, q_conformal, q_experiment],
        yerr=[0, 0, q_error], color=['gray', 'red', 'blue'])
ax2.set_ylabel('Quality Factor (Q)')
ax2.set_title('Q-Factor Enhancement')

plt.tight_layout()
plt.savefig('rlc_resonator.pdf')
# Frequency response
f_classical = 5.03  # MHz
f_conformal = 4.82
f_experiment = 4.85

# Q-factor
q_classical = 15.8
q_conformal = 16.6
q_experiment = 16.4
q_error = 0.3

# Create subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))

# Subplot 1: Frequency shift
ax1.axvline(f_classical, color='k', linestyle=':', label='Classical $f_0$')
ax1.axvline(f_conformal, color='r', linestyle='--', label='Theoretical (This work)')
ax1.axvline(f_experiment, color='b', marker='o', label='Experimental (Paez et al., 2014)')
ax1.set_xlabel('Frequency (MHz)')
ax1.set_title('Resonant Frequency Shift ($\kappa=0.05$)')
ax1.legend()

# Subplot 2: Q-factor
ax2.bar(['Classical', 'This work', 'Experiment'],
        [q_classical, q_conformal, q_experiment],
        yerr=[0, 0, q_error], color=['gray', 'red', 'blue'])
ax2.set_ylabel('Quality Factor (Q)')
ax2.set_title('Q-Factor Enhancement')

plt.tight_layout()
plt.savefig('rlc_resonator.pdf')
```
