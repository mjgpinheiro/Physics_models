```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
S0 = 1
I0 = 1
omega = 2 * np.pi  # Angular frequency for one cycle per unit time
phi = 0  # Phase for entropy
delta = np.pi / 4  # Phase shift between entropy and information

# Time array
t = np.linspace(0, 2, 1000)  # Two cycles

# Entropy S(t)
S = S0 * np.cos(omega * t + phi)

# Information I(t)
I = I0 * np.cos(omega * t + phi + delta)

# Plotting
plt.figure(figsize=(12, 6))

plt.plot(t, S, label='Entropy S(t) = S0 * cos(ωt + φ)')
plt.plot(t, I, label='Information I(t) = I0 * cos(ωt + φ + δ)', linestyle='--')

plt.title('Periodic Dependency of Entropy and Information on Time')
plt.xlabel('Time t')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

# Save the plot as an image file
plt.savefig('Periodic_Dependency_Entropy_Information.png')

# Show the plot
plt.show()
```
